#### module statistics

````python
# @title
import statistics

sample = [10, 203, 20, 30, 40, 10, 20, 50, 60, 70, 10, 20, 80, 100]

# o valor NaN (Not a Number) afecta o comportamento de muitas destas funções,
# ou seja, convém remover os NaN das listas antes de invocar estas funções
res = statistics.mean(sample)
print("Média: ", res)
# fmean -> mais rápido, converte todos os valores para float

res = statistics.median(sample)
print("Mediana: ", res)
#res = statistics.median_low(data)
#res = statistics.median_high(data)

res = statistics.stdev(sample)
print("Devsio padrão: ", res)
# pstdev - toda a população

res = statistics.mode(sample)
print("Moda: ", res)

res = statistics.multimode(sample)
print("Modas (por ordem de aparecimento na lista): ", res)

res = statistics.variance(sample)
print("Variância da amostra:", res)
# pvariance - toda a população
````

````python
# @title exemplo de switch/case - preparação para o shuffle()

# preparação do baralho de cartas
naipes = ["Paus", "Copas", "Ouros", "Espadas"]

def codigo_da_carta(nome_do_naipe, posicao_no_naipe):

  face_da_carta = '?'

  match posicao_no_naipe:
      case 1:
          face_da_carta = "A"

      case 11:
          face_da_carta = "J"

      case 12:
          face_da_carta = "Q"

      case 13:
          face_da_carta = "K"
      case _:
          face_da_carta = str(posicao_no_naipe)

  return face_da_carta + nome_do_naipe[0]

baralho = []
for naipe in naipes:
  baralho += list(codigo_da_carta(naipe,i) for i in range(1,14))
print(baralho)

````

````python
# @title baralhar - shuffle()

print(baralho)

# baralhar
from random import shuffle
shuffle(baralho)
print(baralho)
````

### percorrendo dicionários

````python
# @title items()
populacao = {
    "Portugal": 10196709.7,
    "Espanha": 46733038.9,
    "México": 126190788.0,
    "Egito": 100549000.0
}

for k, v in populacao.items():
    print(k, '{:,}'.format(v), sep=": ")
````

````python
# @title keys()
populacao = {
    "Portugal": 10196709,
    "Espanha": 46733038,
    "México": 126190788,
    "Egito": 100549000
}

print("keys")
paises = populacao.keys()

for pais in paises:
  print(pais)
````

````python
# @title values()
populacao = {
    "Portugal": 10196709,
    "Espanha": 46733038,
    "México": 126190788,
    "Egito": 100549000
}

print ("\nvalues")
habitantes_por_pais = populacao.values()
for numero_de_habitantes in habitantes_por_pais:
  print(numero_de_habitantes)
````

````python
# @title zip()
populacao = {
    "Portugal": 10196709,
    "Espanha": 46733038,
    "México": 126190788,
    "Egito": 100549000
}

paises = populacao.keys()
habitantes_por_pais = populacao.values()

populacao_zip = zip(paises, habitantes_por_pais)
````

````python
# @title zip() - percorendo o iterador

for item in populacao_zip:
  print(item[0], '{:,}'.format(item[1]), sep=": ")
````

````python
# @title zip() - percorendo o iterador (alternativo)

# zip() é um iterador, só pode ser percorrido uma vez.

#populacao_zip2 = zip(paises, habitantes_por_pais)
for pais, numero_de_habitantes in populacao_zip:
  print(pais,'{:,}'.format(numero_de_habitantes), sep=": ")
````

````python
# @title zip() - percorrer várias vezes

# se o convertermos em list pode-se percorrer a lista muitas vezes

populacao_zip3 = list(zip(paises, habitantes_por_pais))
print(populacao_zip3, end="\n\n")

for par in populacao_zip3:
  print(par[0], '{:,}'.format(par[1]), sep=": ")

print("\nrepetindo (usando o ciclo alternativo)...")

for pais, numero_de_habitantes in populacao_zip3:
  print(pais, '{:,}'.format(numero_de_habitantes), sep=": ")
````

````python
# @title obter detalhes de uma entrada
super_heroi = "Mulher-Maravilha"

if (super_heroi not in super_herois):
  print("A personagem %s não está no dicionário de super-heróis"
        % super_heroi),
else:
  print(f"A personagem '{super_heroi}' é um super-herói!",
        "detalhes:",
        sep="\n")

  for detalhe_nome, detalhe_valor in super_herois[super_heroi].items():
    print(f" - {detalhe_nome}", detalhe_valor, sep=": ")
````
