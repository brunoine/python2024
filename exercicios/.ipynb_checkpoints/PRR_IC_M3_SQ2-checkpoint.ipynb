```python
# verifica contagens para a coluna DTCC_COD
df_hosp['DTCC_COD_COUNT']= df_hosp.DTCC_COD.map(df_hosp.DTCC_COD.value_counts(normalize=True))

# Constroi a nova coluna
# Correr sem criar a coluna primeiro
# df_hosp["DTCC_COD_NEW"] = ""
df_hosp.loc[df_hosp['DTCC_COD_COUNT'] < 0.1, 'DTCC_COD_NEW'] = 'outro'
df_hosp.loc[df_hosp['DTCC_COD_COUNT'] >= 0.1, 'DTCC_COD_NEW'] = df_hosp['CC_DSG']
```

### 5.3.3 ImputaÃ§Ã£o de VariÃ¡veis CategÃ³ricas
AlÃ©m do SimpleImputer com a estratÃ©gia 'constant' podemos simplesmente atribuir um valor constante usando um mÃ©todo semelhante ao visto antes, agora para todas as colunas categÃ³ricas

```python
# Seleciona todas as colunas categÃ³ricas e atribui-lhes 
# um novo nÃ­vel criado 'missing'
cols_categorical = df_hosp.select_dtypes(include=['object']).columns.tolist()

df_hosp[cols_categorical] = df_hosp[cols_categorical].fillna('missing')
print(cols_categorical)
```

um Ãºltimo exemplo em que usamos vÃ¡rios imputadores em simultÃ¢neo, seria possÃ­vel usando o ColumnTransformer do mÃ³dulo sklearn.compose e especificando as colunas a fectar Ã© apresentado abaixo:

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

A = [[np.nan,2,'Jose'],[4,np.nan,'Maria'],[7,8,'Joao'],[10,5,np.nan]]

col_trans = ColumnTransformer(
[('imp0', SimpleImputer(strategy='constant', fill_value=1), [0]),
 ('imp1', SimpleImputer(strategy='mean'), [1]),
 ('imp2', SimpleImputer(strategy='constant', fill_value='desconhecido'), [2])],
remainder='passthrough')

col_trans.fit_transform(A)
```

## 5.5 VariÃ¡veis Correlacionadas
Quando as variÃ¡veis no nosso conjunto de dados sÃ£o altamente correlacionadas, podemos ter um problema de multicolinearidade. <br>VariÃ¡veis altamente correlacionadas com uma ou mais variÃ¡veis num modelo de regressÃ£o terÃ£o coeficientes instÃ¡veis e que variarÃ£o significativamente de uma amostra para outra. <br><br>
Ou seja, em modelos baseados em regressÃ£o, a correlaÃ§Ã£o entre variÃ¡veis pode ser problemÃ¡tica porque cria modelos instÃ¡veis onde a inferÃªncia estatÃ­stica se torna difÃ­cil ou pouco confiÃ¡vel.

```python
# criaÃ§Ã£o de matriz de correlaÃ§Ã£o e selecÃ£o do triÃ¢ngulo superior
cor_matrix = df_x.corr().abs()
cor_matrix
```

```python
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(bool))
print(upper_tri)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# o heatmap Ã© uma boa forma de visualizar as matrizes de correlaÃ§Ãµes
plt.figure(figsize = (20,20))
sns.heatmap(cor_matrix)
```

```python
# seleciona para remover as colunas altamente correlacionadas
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]
print(to_drop)
```

```python
df_hosp3 = df_x.drop(columns=to_drop, axis=1)
df_hosp3.head()
```

```python
import numpy.random as rd
import pandas as pd
from sklearn.preprocessing import Normalizer
import seaborn as sns

import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")

shape = (100, 2)
# *shape faz unpack do tuplo shape
df = pd.DataFrame(rd.rand(*shape) * rd.lognormal(1, 0.4, shape)
                  , columns=["weight", "age"])
ndf = pd.DataFrame(Normalizer(norm="l2").fit_transform(df),
                   columns=["norm_weight", "norm_age"])

# kernel density estimate (KDE) plot Ã© um mÃ©todo para 
# visualizar a distribuiÃ§Ã£o de observaÃ§Ãµes num dataset
sns.kdeplot(data=pd.concat([df, ndf], axis=1), fill=True, 
            common_norm=False, palette="crest",alpha=.5, linewidth=1,)
```

```python
for d in [df]:
    sns.pairplot(d.reset_index(), hue="index", diag_kind=None)
```

```python
import numpy.random as rd
import pandas as pd
from sklearn.preprocessing import StandardScaler
import seaborn as sns

shape = (50, 2)
# *shape faz unpack do tuplo shape
df = pd.DataFrame(rd.rand(*shape) * rd.normal(10, 6, shape),
                  columns=["weight", "age"])
ndf = pd.DataFrame(StandardScaler().fit_transform(df),
                   columns=["norm_weight", "norm_age"])

# kernel density estimate (KDE) plot Ã© um mÃ©todo para 
# visualizar a distribuiÃ§Ã£o de observaÃ§Ãµes num dataset
sns.kdeplot(data=pd.concat([df, ndf], axis=1), 
            fill=True, common_norm=False, palette="crest",
            alpha=.5, linewidth=1,)
```

```python
for d in [df]:
    sns.pairplot(d.reset_index(), hue="index", palette="crest", diag_kind=None)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold # Feature selector
from sklearn.pipeline import Pipeline # For setting up pipeline
from sklearn.neighbors import KNeighborsClassifier

#Define a pre-processing pipeline
pipeline = Pipeline([
('selector', VarianceThreshold()),
('scaler', StandardScaler()),
('KNN', KNeighborsClassifier())])
```
