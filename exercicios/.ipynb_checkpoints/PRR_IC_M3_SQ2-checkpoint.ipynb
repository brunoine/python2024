Algumas das variáveis categóricas no nosso conjunto de dados podem possuir centenas de classes diferentes. Isto pode ser problemático se a amostra selecionada para treinar o modelo não abranger todas as combinações possíveis. <br>
A melhor maneira de lidar com este problema é aumentar o tamanho da amostra. Quando tal não é uma opção podemos agregar categorias.<br><br>

Abaixo agregamos todos os municipios com menos de dez por cento das ocorrências na amostra. 

```python
# verifica contagens para a coluna DTCC_COD
df_hosp['DTCC_COD_COUNT']= df_hosp.DTCC_COD.map(df_hosp.DTCC_COD.value_counts(normalize=True))

# Constroi a nova coluna
# Correr sem criar a coluna primeiro
# df_hosp["DTCC_COD_NEW"] = ""
df_hosp.loc[df_hosp['DTCC_COD_COUNT'] < 0.1, 'DTCC_COD_NEW'] = 'outro'
df_hosp.loc[df_hosp['DTCC_COD_COUNT'] >= 0.1, 'DTCC_COD_NEW'] = df_hosp['CC_DSG']
```

```python
# exclusão das colunas já tratadas
df_hosp = df_hosp.drop(columns=['DTCC_COD', 'CC_DSG','DTCC_COD_COUNT'])
df_hosp.head()
```

### 5.3.3 Imputação de Variáveis Categóricas
Além do SimpleImputer com a estratégia 'constant' podemos simplesmente atribuir um valor constante usando um método semelhante ao visto antes, agora para todas as colunas categóricas

```python
# Seleciona todas as colunas categóricas e atribui-lhes 
# um novo nível criado 'missing'
cols_categorical = df_hosp.select_dtypes(include=['object']).columns.tolist()

df_hosp[cols_categorical] = df_hosp[cols_categorical].fillna('missing')
print(cols_categorical)
```

um último exemplo em que usamos vários imputadores em simultâneo, seria possível usando o ColumnTransformer do módulo sklearn.compose e especificando as colunas a fectar é apresentado abaixo:

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

A = [[np.nan,2,'Jose'],[4,np.nan,'Maria'],[7,8,'Joao'],[10,5,np.nan]]

col_trans = ColumnTransformer(
[('imp0', SimpleImputer(strategy='constant', fill_value=1), [0]),
 ('imp1', SimpleImputer(strategy='mean'), [1]),
 ('imp2', SimpleImputer(strategy='constant', fill_value='desconhecido'), [2])],
remainder='passthrough')

col_trans.fit_transform(A)
```

### 5.4.1 Criação de Variáveis Categóricas
podemos adicionar novas colunas ou mudar colunas da nossa dataframe, usando uma series do pandas para o fazer. <br>
Mas se a conseguirmos criar à custa de outras colunas podemos fazê-lo directamente usando o método **map**

```python
def label(value):
    if value == 0:
        return "no"
    if value > 0:
        return "yes"

df_hosp2['t_cirurgia'] = df_hosp2['C21071'].map(label)
df_hosp2.head()
```

## 5.5 Variáveis Correlacionadas
Quando as variáveis no nosso conjunto de dados são altamente correlacionadas, podemos ter um problema de multicolinearidade. <br>Variáveis altamente correlacionadas com uma ou mais variáveis num modelo de regressão terão coeficientes instáveis e que variarão significativamente de uma amostra para outra. <br><br>
Ou seja, em modelos baseados em regressão, a correlação entre variáveis pode ser problemática porque cria modelos instáveis onde a inferência estatística se torna difícil ou pouco confiável.

```python
# criação de matriz de correlação e selecão do triângulo superior
cor_matrix = df_x.corr().abs()
cor_matrix
```

```python
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(bool))
print(upper_tri)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# o heatmap é uma boa forma de visualizar as matrizes de correlações
plt.figure(figsize = (20,20))
sns.heatmap(cor_matrix)
```

```python
# seleciona para remover as colunas altamente correlacionadas
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]
print(to_drop)
```

```python
df_hosp3 = df_x.drop(columns=to_drop, axis=1)
df_hosp3.head()
```

## 5.6 Class Imbalance
Normalmente, a subamostragem/sobreamostragem será feita apenas nos dados de treino, esta é a abordagem correta.<br>
Ao usar a biblioteca python sklearn para treinar o classificador, define-se o parâmetro *class_weight='balanced'*.

Se se fizer a subamostragem antes da partição em conjunto de treino e de teste, a distribuição deste último poderá não replicar a distribuição dos dados do mundo real. Conseqüentemente, normalmente evita-se a amostragem antes da partição.

```python
from sklearn.linear_model import LogisticRegression

# exemplo de declaração do regressor balanceando as classes
Lr = LogisticRegression(class_weight='balanced')
```

```python
import numpy.random as rd
import pandas as pd
from sklearn.preprocessing import Normalizer
import seaborn as sns

import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")

shape = (100, 2)
# *shape faz unpack do tuplo shape
df = pd.DataFrame(rd.rand(*shape) * rd.lognormal(1, 0.4, shape)
                  , columns=["weight", "age"])
ndf = pd.DataFrame(Normalizer(norm="l2").fit_transform(df),
                   columns=["norm_weight", "norm_age"])

# kernel density estimate (KDE) plot é um método para 
# visualizar a distribuição de observações num dataset
sns.kdeplot(data=pd.concat([df, ndf], axis=1), fill=True, 
            common_norm=False, palette="crest",alpha=.5, linewidth=1,)
```

```python
import numpy.random as rd
import pandas as pd
from sklearn.preprocessing import StandardScaler
import seaborn as sns

shape = (50, 2)
# *shape faz unpack do tuplo shape
df = pd.DataFrame(rd.rand(*shape) * rd.normal(10, 6, shape),
                  columns=["weight", "age"])
ndf = pd.DataFrame(StandardScaler().fit_transform(df),
                   columns=["norm_weight", "norm_age"])

# kernel density estimate (KDE) plot é um método para 
# visualizar a distribuição de observações num dataset
sns.kdeplot(data=pd.concat([df, ndf], axis=1), 
            fill=True, common_norm=False, palette="crest",
            alpha=.5, linewidth=1,)
```
