### Parágrafos


**Parágrafos e Novas Linhas**
Estas linhas 
são apresentadas na mesma linha


A seguir de **2 ou mais SPACES** o texto continua na próxima  linha  
Estas linhas já não

Em alternativa utilizar HTML tag \<br> <br>
para continuar na próxima linha. Estes tags são mais visiveis que "2 ou mais SPACES"

Carateres reservados podem ser incluidos utilizando uma barra invertida \, por exemplo par amostrar \*\*Este texto não está em BOLD\*\*
