#### max()

````python
# @title nmeros
# para um conjunto de elementos, retorna o maior

var1 = 4
var2 = 8
var3 = 2

max_val = max(var1, var2, var3)

print(max_val)
````

````python
# @title alfabeticamente maior

var1 = "amostra"
var2 = "universo"
var3 = "ndice"

max_val = max(var1, var2, var3)

print(max_val)
````

````python
# @title mais comprida
var1 = "sim"
var2 = "anticonstitucionalissimamente"
var3 = "Pneumoultramicroscpicossilicovulcanoconitico"

max_val = max(var1, var2, var3,
            key=len)

print(max_val)
````

````python
# @title maior letra

string = "Vero Azul"
max_val = max(string)
print(ord('A'),ord('a'), ord(''))
print(max_val)
````

````python
# @title alfabeticamente maior (de uma lista)

string_list = ["Azul", "p", "estatstica"]

max_val = max(string_list)
print(max_val)

# a mais comprida
max_val = max(string_list, key=len)
print(max_val)
````

#### module statistics

````python
# @title
import statistics

sample = [10, 203, 20, 30, 40, 10, 20, 50, 60, 70, 10, 20, 80, 100]

# o valor NaN (Not a Number) afecta o comportamento de muitas destas funes,
# ou seja, convm remover os NaN das listas antes de invocar estas funes
res = statistics.mean(sample)
print("Mdia: ", res)
# fmean -> mais rpido, converte todos os valores para float

res = statistics.median(sample)
print("Mediana: ", res)
#res = statistics.median_low(data)
#res = statistics.median_high(data)

res = statistics.stdev(sample)
print("Devsio padro: ", res)
# pstdev - toda a populao

res = statistics.mode(sample)
print("Moda: ", res)

res = statistics.multimode(sample)
print("Modas (por ordem de aparecimento na lista): ", res)

res = statistics.variance(sample)
print("Varincia da amostra:", res)
# pvariance - toda a populao
````

````python
# @title exemplo de switch/case - preparao para o shuffle()

# preparao do baralho de cartas
naipes = ["Paus", "Copas", "Ouros", "Espadas"]

def codigo_da_carta(nome_do_naipe, posicao_no_naipe):

  face_da_carta = '?'

  match posicao_no_naipe:
      case 1:
          face_da_carta = "A"

      case 11:
          face_da_carta = "J"

      case 12:
          face_da_carta = "Q"

      case 13:
          face_da_carta = "K"
      case _:
          face_da_carta = str(posicao_no_naipe)

  return face_da_carta + nome_do_naipe[0]

baralho = []
for naipe in naipes:
  baralho += list(codigo_da_carta(naipe,i) for i in range(1,14))
print(baralho)

````

````python
# @title baralhar - shuffle()

print(baralho)

# baralhar
from random import shuffle
shuffle(baralho)
print(baralho)
````

### percorrendo dicionrios

````python
# @title items()
populacao = {
    "Portugal": 10196709.7,
    "Espanha": 46733038.9,
    "Mxico": 126190788.0,
    "Egito": 100549000.0
}

for k, v in populacao.items():
    print(k, '{:,}'.format(v), sep=": ")
````

````python
# @title keys()
populacao = {
    "Portugal": 10196709,
    "Espanha": 46733038,
    "Mxico": 126190788,
    "Egito": 100549000
}

print("keys")
paises = populacao.keys()

for pais in paises:
  print(pais)
````

````python
# @title values()
populacao = {
    "Portugal": 10196709,
    "Espanha": 46733038,
    "Mxico": 126190788,
    "Egito": 100549000
}

print ("\nvalues")
habitantes_por_pais = populacao.values()
for numero_de_habitantes in habitantes_por_pais:
  print(numero_de_habitantes)
````

````python
# @title zip()
populacao = {
    "Portugal": 10196709,
    "Espanha": 46733038,
    "Mxico": 126190788,
    "Egito": 100549000
}

paises = populacao.keys()
habitantes_por_pais = populacao.values()

populacao_zip = zip(paises, habitantes_por_pais)
````

````python
# @title zip() - percorendo o iterador

for item in populacao_zip:
  print(item[0], '{:,}'.format(item[1]), sep=": ")
````

````python
# @title zip() - percorendo o iterador (alternativo)

# zip()  um iterador, s pode ser percorrido uma vez.

#populacao_zip2 = zip(paises, habitantes_por_pais)
for pais, numero_de_habitantes in populacao_zip:
  print(pais,'{:,}'.format(numero_de_habitantes), sep=": ")
````

````python
# @title zip() - percorrer vrias vezes

# se o convertermos em list pode-se percorrer a lista muitas vezes

populacao_zip3 = list(zip(paises, habitantes_por_pais))
print(populacao_zip3, end="\n\n")

for par in populacao_zip3:
  print(par[0], '{:,}'.format(par[1]), sep=": ")

print("\nrepetindo (usando o ciclo alternativo)...")

for pais, numero_de_habitantes in populacao_zip3:
  print(pais, '{:,}'.format(numero_de_habitantes), sep=": ")
````

### eliminar

````python
# @title remove por chave (se a chave no existir, d erro)
d = {'a': 1, 'b': 2, 'c':3}
print('dicionrio', d, sep="\n")

k = "b"

eliminado = d.pop(k)

print(f"\nA entrada '{k}' tinha o valor '{eliminado}' e foi removida",
      d, sep='\n')
````

````python
# @title remove por chave (se a chave no existir, devolve o valor default)
d = {'a': 1, 'b': 2, 'c':3}
print('dicionrio', d, sep="\n")

k = 'x'
default = 99

eliminado = d.pop(k, default)

print(f"\nA entrada '{k}' tinha o valor '{eliminado}' (no existia no dicionrio)",
      d, sep='\n')
````

### dicionrios com outros objetos

````python
# @title
super_herois = {
    "Super-Homem": {
        "Habilidade Especial": "Super Fora e Voo",
        "Poder": 10,
        "Editora": "DC Comics"
    },
    "Homem-Aranha": {
        "Habilidade Especial": "Agilidade e Sentido Aranha",
        "Poder": 8,
        "Editora": "Marvel Comics"
    },
    "Mulher-Maravilha": {
        "Habilidade Especial": "Fora, Velocidade e Luta",
        "Poder": 9,
        "Editora": "DC Comics"
    },
    "Homem de Ferro": {
        "Habilidade Especial": "Tecnologia e Armadura Avanada",
        "Poder": 7,
        "Editora": "Marvel Comics"
    },
    "Flash": {
        "Habilidade Especial": "Super Velocidade",
        "Poder": 9,
        "Editora": "DC Comics"
    }
}

````

````python
# @title obter detalhes de uma entrada
super_heroi = "Mulher-Maravilha"

if (super_heroi not in super_herois):
  print("A personagem %s no est no dicionrio de super-heris"
        % super_heroi),
else:
  print(f"A personagem '{super_heroi}'  um super-heri!",
        "detalhes:",
        sep="\n")

  for detalhe_nome, detalhe_valor in super_herois[super_heroi].items():
    print(f" - {detalhe_nome}", detalhe_valor, sep=": ")
````

````python
# @title pesquisar dentro do dicionrio interior
def compara_poderes(super_heroi):
  return super_herois[super_heroi]["Poder"]
heroi_mais_poderoso = max(super_herois, key=compara_poderes)
print("\nHeri mais poderoso: %s" % heroi_mais_poderoso)
````

#### 1. Dado um conjunto de 3 users e 3 passwords (armazenados num dicionrio) pea ao utilizador para introduzir um user e verifique se faz parte do dicionrio e de seguida pea a password correspondente e a valide

````python
# @title
users = {
    "ana": "pass1",
    "bruno": "pass2",
    "hugo": "pass3",
    "ines": "pass4",
    "maria": "pass5"
}

user = input("login:")

if (user in users):

  password = input("password:")

  if (users[user]==password):
    print(f"Ol {user}")
  else:
    print("password invlida")
else:
  print("utilizador no existe")
````
